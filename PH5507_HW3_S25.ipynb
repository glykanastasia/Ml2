```python
import numpy as np
import pandas as pd
import tensorflow as tf
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM, GRU, Bidirectional
import matplotlib.pyplot as plt

# PARAMETERS
TICKER = '^GSPC'      # ή π.χ. 'AAPL'
START = '2022-01-01'
END   = '2024-01-01'
SEQ_LEN = 100          # μέγεθος παραθύρου (100 ημέρες)
EPOCHS = 30
BATCH_SIZE = 64
MODELS = ['SimpleRNN', 'LSTM', 'GRU']
HORIZONS = {'1-day':1, '7-day':7}
UNITS = 100
LATENT_DIM = 10         # για VAE

# --- Part 1: VAE for Synthetic Series Generation ---
# 1. Load data and compute returns
data = yf.download(TICKER, start=START, end=END)
price_col = 'Adj Close' if 'Adj Close' in data.columns else 'Close'
prices = data[price_col]
returns = prices.pct_change().dropna().values.reshape(-1,1)
# 2. Normalize returns to [0,1]
min_val, max_val = returns.min(), returns.max()
returns_norm = (returns - min_val) / (max_val - min_val)
# 3. Create sequences of length SEQ_LEN
def create_sequences_simple(series, window):
    X = []
    for i in range(len(series) - window):
        X.append(series[i:i+window])
    return np.array(X)
X_all = create_sequences_simple(returns_norm, SEQ_LEN)

# Sampling layer for VAE
def Sampling():
    class _Sampling(layers.Layer):
        def call(self, inputs):
            z_mean, z_log_var = inputs
            batch = tf.shape(z_mean)[0]
            dim = tf.shape(z_mean)[1]
            epsilon = tf.random.normal(shape=(batch, dim))
            return z_mean + tf.exp(0.5 * z_log_var) * epsilon
    return _Sampling()

# Encoder
encoder_inputs = keras.Input(shape=(SEQ_LEN,1))
x = layers.Flatten()(encoder_inputs)
x = layers.Dense(64, activation='relu')(x)
z_mean = layers.Dense(LATENT_DIM)(x)
z_log_var = layers.Dense(LATENT_DIM)(x)
z = Sampling()([z_mean, z_log_var])
encoder = keras.Model(encoder_inputs, [z_mean, z_log_var, z], name='encoder')

# Decoder
latent_inputs = keras.Input(shape=(LATENT_DIM,))
x = layers.Dense(64, activation='relu')(latent_inputs)
x = layers.Dense(SEQ_LEN, activation='sigmoid')(x)
decoder_outputs = layers.Reshape((SEQ_LEN,1))(x)
decoder = keras.Model(latent_inputs, decoder_outputs, name='decoder')

# VAE Model
class VAE(keras.Model):
    def __init__(self, encoder, decoder, **kwargs):
        super().__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
    def compile(self, optimizer, loss_fn):
        super().compile()
        self.optimizer = optimizer
        self.loss_fn = loss_fn
        self.total_loss_tracker = keras.metrics.Mean(name='total_loss')
        self.reconstruction_loss_tracker = keras.metrics.Mean(name='reconstruction_loss')
        self.kl_loss_tracker = keras.metrics.Mean(name='kl_loss')
    @property
    def metrics(self):
        return [self.total_loss_tracker,
                self.reconstruction_loss_tracker,
                self.kl_loss_tracker]
    def train_step(self, data):
        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data)
            reconstruction = self.decoder(z)
            rec_loss = tf.reduce_mean(self.loss_fn(data, reconstruction)) * SEQ_LEN
            kl_loss = -0.5 * tf.reduce_mean(
                tf.reduce_sum(1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var), axis=1)
            )
            total_loss = rec_loss + kl_loss
        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        self.total_loss_tracker.update_state(total_loss)
        self.reconstruction_loss_tracker.update_state(rec_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        return {'loss': self.total_loss_tracker.result(),
                'reconstruction_loss': self.reconstruction_loss_tracker.result(),
                'kl_loss': self.kl_loss_tracker.result()}

# Train VAE
vae = VAE(encoder, decoder)
vae.compile(optimizer=keras.optimizers.Adam(),
            loss_fn=keras.losses.MeanSquaredError())
vae.fit(X_all, epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=1)

# Generate synthetic sequences
tf.random.set_seed(42)
codings = tf.random.normal(shape=(8, LATENT_DIM))
generated = decoder.predict(codings)
# Plot synthetic
fig, axes = plt.subplots(2, 4, figsize=(16,6))
axes = axes.flatten()
for i, ax in enumerate(axes):
    ax.plot(generated[i])
    ax.set_title(f'Synthetic {i+1}')
plt.tight_layout()
plt.show()

# --- Part 2: RNN Models for Next-Day & 7-Day Prediction ---
# 1. Prepare train/test sequences
# Reuse returns_norm from above

def create_xy(series, seq_len, horizon):
    X, y = [], []
    for i in range(len(series) - seq_len - horizon + 1):
        X.append(series[i:i+seq_len])
        y.append(series[i+seq_len+horizon-1])
    return np.array(X), np.array(y)

datasets = {}
for name, h in HORIZONS.items():
    X, y = create_xy(returns_norm, SEQ_LEN, h)
    split = int(0.8 * len(X))
    datasets[name] = (X[:split], y[:split], X[split:], y[split:])

# 2. Model builder
def build_model(model_type, units, seq_len, bidirectional=False):
    model = Sequential()
    if model_type == 'SimpleRNN':
        layer = SimpleRNN(units, input_shape=(seq_len,1))
    elif model_type == 'LSTM':
        layer = LSTM(units, input_shape=(seq_len,1))
    elif model_type == 'GRU':
        layer = GRU(units, input_shape=(seq_len,1))
    if bidirectional:
        layer = Bidirectional(layer)
    model.add(layer)
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model

# 3. Train & Evaluate
histories = {}
predictions = {}
for m in MODELS:
    for name, (X_tr, y_tr, X_te, y_te) in datasets.items():
        key = f"{m}_{name}"
        bi = (m in ['LSTM','GRU'])
        model = build_model(m, UNITS, SEQ_LEN, bidirectional=bi)
        hist = model.fit(X_tr, y_tr,
                         epochs=EPOCHS,
                         batch_size=BATCH_SIZE,
                         validation_split=0.1,
                         verbose=1)
        pred = model.predict(X_te)
        pred_inv = scaler.inverse_transform(pred)
        true_inv = scaler.inverse_transform(y_te.reshape(-1,1))
        rmse = np.sqrt(mean_squared_error(true_inv, pred_inv))
        print(f"{key} RMSE: {rmse:.6f}")
        histories[key] = hist
        predictions[key] = (pred_inv, true_inv)

# 4. Plot best prediction
best = min(predictions, key=lambda k: np.sqrt(mean_squared_error(predictions[k][1], predictions[k][0])))
pred_best, true_best = predictions[best]
plt.figure(figsize=(14,6))
plt.plot(true_best, label='Actual')
plt.plot(pred_best, label='Predicted')
plt.title(f'Best Model: {best}')
plt.legend()
plt.show()
```
